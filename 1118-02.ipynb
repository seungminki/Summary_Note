# 과제 - 임의의 시각화 및 분석 해보기

```python
df_jeju = pd.read_csv("C:/Users/user/Downloads/한국전력거래소_제주 월별 발전기별 전력거래량_20211231.csv", encoding='cp949')
```

```python
# 2015년 기준으로 이전, 이후 전력거래량의 변화 비교
```

```python
before_15 = df_jeju.iloc[:,:49].mean()
```

```python
after_15 = df_jeju.iloc[:,49:].mean()
```

```python
before_15
```

```python
after_15
```

```python
import seaborn as sns
```

```python
sns.kdeplot(before_15)
```

```python
sns.kdeplot(after_15)
```

```python
stats.ks_2samp(after_15, before_15)
```

```python
stats.probplot(before_15, plot=plt)
plt.show()
```

```python
stats.probplot(after_15, plot=plt)
plt.show()
```

```python
stats.shapiro(before_15)
```

```python
stats.shapiro(after_15)
```

```python
before_15 = before_15.reset_index(drop=True)
after_15 = after_15.reset_index(drop=True)
```

```python
before_15
```

```python
after_15
```

```python
total_data = pd.concat([before_15, after_15], axis=1, ignore_index=True) # concat(데이터 합치는 함수)
```

```python
total_data
```

```python
total_data = pd.melt(total_data) # melt: 컬럼을 녹여서 행으로 보낸다
```

```python
total_data
```

```python
total_data['variable'] = total_data['variable'].apply(lambda x: 'before' if x == 0 else 'after')
```

```python
total_data
```

```python
from statannot import add_stat_annotation
```

```python
ax = plt.subplot()
ax = sns.boxplot(
    data=total_data, 
    x='variable', 
    y='value')

# ax = figure 이름
# data = 분석할 데이터 (여기서는 데이터프레임 형식)
# x = x축에 위치할 정보 (여기서는 column 이름)
# y = y축에 위치할 정보 (여기서는 column 이름)
# box_pairs = 분석할 박스플롯의 쌍을 입력 (리스트로 입력받음, 한 쌍마다 튜플로 묶어둠)
# test = 분석할 T-test의 종류. 모를 경우 여기에 아무거나 입력하고 에러를 발생시키면 가능한 경우를 알려줌
# text_format = T-test 결과를 별 표시로 알려줌. return 창에 관련 정보를 함께 띄워줌
# loc = 도표 내부에 표시할건지 외부에 표시할건지
# verbose = return 창에 관련 설명을 어디까지 표시할건지
add_stat_annotation(ax, 
                    data=total_data, 
                    x = 'variable', 
                    y = 'value',
                    box_pairs=[(('before'), ('after'))],
                    test='t-test_ind', 
                    text_format='star', 
                    loc='inside', 
                    verbose=2)
```
