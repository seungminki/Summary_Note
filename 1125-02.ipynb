```python
# 문자 처리를 위한 패키지 re
import re
```

```python
# seach 함수는 처음 문자열에 해당하는 항목을 두 번째 문자열에서 찾는다
# 하나를 찾으면 종결
m = re.search('[0-9]', '010-1234-4567')
```

```python
# 대상은 group 함수로 찾아볼 수 있다
m.group()
```

```python
# 정규식의 변화에 따른 search 함수의 동작 차이 알아보기
m = re.search('[0-9]+', '010-1234-4567')
```

```python
m.group()
```

```python
# search 함수의 parameter: 찾을 문자, 대상 문자
m = re.search('o', 'hello, world!')
```

```python
m.group()
```

## 정규식
#### 특정 문자의 집합을 알기 쉽게 표현한 것
1. 숫자 [0-9] : 숫자 전체 (단, 한 자리만)
1. 숫자 [0-9]+ : 숫자 전체 (자릿수 제한 없음)
1. 영어 [a-z] : 영어 소문자 전체 (단, 한 문자만)
1. 영어 [a-z]+ : 영어 소문자 전체 (문자 수 제한 없음)
1. 영어 [A-Z] : 영어 대문자 전체 (단, 한 문자만)
1. 영어 [A-Z] : 영어 데문자 전체 (문자 수 제한 없음)
1. 한글 [ㄱ-ㅎ] : 한글 자음 전체 (단, 한 문자만)
1. 한글 [ㄱ-ㅎ]+ : 한글 자음 전체 (문자 수 제한 없음)
1. 한글 [ㅏ-ㅣ] : 한글 모음 전체 (단, 한 문자만)
1. 한글 [ㅏ-ㅣ]+ : 한글 모음 전체 (문자 수 제한 없음)
1. 한글 [가-하] : 한글 전체 (단, 한 문자만) - 복합모음 제외
1. 한글 [가-하]+ : 한글 전체 (문자 수 제한 없음) - 복합모음 제외
1. 한글 [가-힣]+ : 한글 전체 (문자 수 제한 없음)
1. 복합 [a-z|가-힣]+ : 영어 소문자, 한글 전체 (문자 수 제한 없음)
1. 특문 [ ]+ : 띄어쓰기 (문자 수 제한 없음, 다른 특수문자 조합 가능)

```python
def time_parse(x):
    x = re.findall('[\d]+:[\d]+:[\d]+',str(x))[0]
    x = x.split(':')
    x = f'{x[0]}시{x[1]}분{x[2]}초'
    
    return x

df['주문일자'].apply(lambda x: time_parse(x))
```

## 실습
* 상기 데이터프레임에서 그룹별 통계분석을 해 보기
* 데이터쌍은 임의로 선정

```python
m = est['대장금액(원)'] > 50000
est2 = est[m]
m = est2['대장면적(제곱미터)'] > 10
est2 = est2[m]
```

```python
sns.distplot(np.log2(est2['대장면적(제곱미터)']+0.001))
```

```python
stats.probplot(np.log2(est2['대장면적(제곱미터)']+0.001), plot=plt)
```

```python
stats.shapiro(np.log2(est2['대장면적(제곱미터)']+0.001))
```

```python
text = stats.spearmanr(np.log2(est2['대장면적(제곱미터)']+0.001), np.log2(est2['대장금액(원)']+0.001))
```

```python
points = plt.scatter(np.log2(est2['대장면적(제곱미터)']+0.001), np.log2(est2['대장금액(원)']+0.001), 
                     marker='+', alpha=0.1)

sns.regplot(np.log2(est2['대장면적(제곱미터)']+0.001), np.log2(est2['대장금액(원)']+0.001), 
            scatter=False, color="r")

plt.text(x=15, y=37, s='r=%.4f, p=%.4f' %(text[0], text[1]))
```
