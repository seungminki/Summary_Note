```python
# 특정 조건 파일 옮기기
# CMK, CKS
def file_move(path, to_dir, word):
    for root, subdirs, files in os.walk(path):
        for f in files:
            if word in f:
                file_to_move = os.path.join(root, f)
                shutil.move(file_to_move, to_dir) 
                print('done')
            else:
                print('fail')
```

```python
path = "D:/test_db/CKS_000-0101-korean/wav"
File_List = glob.glob(path + "\\*.wav")
    
fg = open("D:/test_db/CKS_000-0101-korean/etc/durninfo.txt", 'w')

for i in File_List:
    
    with contextlib.closing(wave.open(i,'r')) as f:
        frames = f.getnframes()
        rate = f.getframerate()
        duration = frames / float(rate)
    
    print(i + " " + str(duration), file=fg)
f.close()
```
