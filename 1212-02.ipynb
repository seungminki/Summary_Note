#### 1. 정규식(Regex)
> 정규식(Regular expressions or regex)은 연속적인 문자들을 검색하는 제어방법입니다. (모질라 개발자 네트워크, MDN 용어 사전)

1) 영문 대소문자 또는 숫자만 포함된 문장인지 검사하는 함수를 작성하라

```python
def is_allowed_specific_char(string):
    charRe = re.compile(r'[^a-zA-Z0-9]') # 정규식 패턴 작성
    string = charRe.search(string)
    return not bool(string) # 결과 전달 방식 주의

print(is_allowed_specific_char("ABCDEFabcdef123450")) # True 
print(is_allowed_specific_char("*&%@#!}{")) # False 
print(is_allowed_specific_char("ABCDEFabcdef123450!")) # False 
```

2) 영문자 a를 포함하며 a뒤에 0개 이상의 b가 이어지는지 검사하라.

```python
# Solution
def text_match(text):
        patterns = 'ab*' # 정규식 패턴 작성성 
        if re.search(patterns,  text):
                return 'Found a match!'
        else:
                return('Not matched!')

print(text_match("bc")) # no match
print(text_match("ac")) # match 
print(text_match("abc")) # match
print(text_match("abbc")) # match
```

7) 영문 소문자 문자열들이 언더바로 이어져 있는지 검사.

```python
# Solution
def text_match(text):
        patterns = '[a-z]+_[a-z]+$' # 정규식 패턴 작성
        if re.search(patterns,  text):
                return 'Found a match!'
        else:
                return('Not matched!')

print(text_match("aab_cbbbc")) # true
print(text_match("aab_Abbbc")) # false
print(text_match("Aaab_abbbc")) # true
print(text_match("aab_cbbbc_")) # false
```

9) 문자열에 a가 있고, a뒤에 문자들이 이어지다가, b로 끝나는지 검사

```python
# Solution
def text_match(text):
        patterns = 'a.*?b$' # 정규식 패턴 작성
        if re.search(patterns,  text):
                return 'Found a match!'
        else:
                return('Not matched!')

print(text_match("aabbbbd")) # false
print(text_match("aabAbbbc")) # false
print(text_match("accddbbjjjb")) # True
```

15) 특정 숫자로 시작하는지 검사. 

```python
# Solution
def match_num(string):
    text = re.compile(r"^5") # 5로 시작하는지 검사하도록 패턴 작성
    if text.match(string):
        return True
    else:
        return False
print(match_num('5-2345861'))
print(match_num('6-2345861'))
```

17) 문자열 끝이 숫자인지 확인하라

```python
# Solution
def end_num(string):
    text = re.compile(r".*[0-9]$")  # 정규식 패턴 작성
    if text.match(string):
        return True
    else:
        return False

print(end_num('abcdef')) # FALSE
print(end_num('abcdef6')) # TRUE
```

20) 문자열에서 특정 문자열을 검색하고, 해당 문자열의 위치를 확인하기

- 문자열 : 'The quick brown fox jumps over the lazy dog.'
- 검색어 : 'fox'

```python
# Solution
pattern = 'fox'
text = 'The quick brown fox jumps over the lazy dog.'
match = re.search(pattern, text)
s = match.start()
e = match.end()
print('Found "%s" in "%s" from %d to %d ' % (match.re.pattern, match.string, s, e))
```

25) 날짜 포멧을 yyyy-mm-dd 에서 dd-mm-yyyy 으로 변경하라.

```python
# Solution
def change_date_format(dt):
        return re.sub(r'(\d{4})-(\d{1,2})-(\d{1,2})', '\\3-\\2-\\1', dt) # 코드 완성
dt1 = "2026-01-02"
print("Original date in YYY-MM-DD Format: ",dt1)
print("New date in DD-MM-YYYY Format: ",change_date_format(dt1))
```
