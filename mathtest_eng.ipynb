```python
import pdfplumber
```

```python
def get_text(page):
    text = page.extract_text()
    return text

def pdf2text(fname):
    pdf = pdfplumber.open(fname)
    pages = pdf.pages
    text = ""
    for pg in pages:
        text += get_text(pg)
    return text
```
