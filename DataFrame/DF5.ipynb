# 데이터와 정보의 차이
* 데이터(data)
    * 단순히 모아놓은 자료의 총칭
    * 별도의 의미가 없는 기록에 그침
    * 데이터는 분석이 불가능하다
    * 기온, 인구수, 방문기록 등

* 정보(information)
    * 데이터를 가공해 특별한 목적을 띤다
    * 분석이 가능한 것이 정보
    * 연 평균 기온, 특정 지역의 전입/전출 기록, 랜드마크별 방문기록

# 데이터의 전처리
* 데이터는 기본적으로 누락, 오차, 형태에 맞지 않는 값 등의 문제가 있다
* -> 데이터 전처리를 통해 해당 요소 제거 가능
    * 데이터 전처리의 종류
    * 데이터 클리닝
    * 데이터 통합
    * 데이터 변환
    * 데이터 축소
    * 데이터 이산화
    * 결측치 및 이상치 처리
    * 일관성 있는 데이터로 변환 (표준화, 정규화)

# 데이터 클리닝
* 중복 데이터 삭제
* 관련 없는 데이터 삭제
* 구조적 문제 수정
    * 대소문자 일치

* 이상치(outlier) 처리

# 분석이 불가하거나 어려운 형태를 가능한 형태로 변환할 수 있다.
* 단위 변환
* (문자열의 경우) 파싱
* 형태 변환
* 여러 의미를 가진 데이터의 경우 분리 저장

# 분석이 불가한 데이터를 저차원, 저용량으로 축소
* 차원 축소
* 유효 숫자 변경
* 데이터 용량 축소
* 수량 감소
    * 패러미터만 남기고 삭제
    * 선형 회귀
    * 히스토그램
    * 클러스터링
    * 샘플링

# 연속형 데이터를 범주형 데이터로 변경
* 기계학습 등에서 주효하게 사용
* 숫자가 크게 높거나 비슷한 값의 연관성 제거
* 컴퓨터가 판단할 수 없다는 특성에 근거

# 분석 불가능하거나 크게 영향을 주는 데이터 처리
* 결측치 처리
    * 제로필
    * 평균값
    * 중간값
    * 기계학습을 통한 대체

* 이상치 처리
    * 제거
    * 대체
    * 기계학습을 통한 대체

# 데이터의 분포에 따른 특성 제거 또는 통계를 위한 변환
* 표준화(정규 분포 형태로 변환)
    * 데이터베이스의 정규화 개념과 구분
    * 데이터의 범위 차이가 달라짐
    * 데이터 범위의 제한 없음
    * Z-score 등

* 정규화(0-1 범위로 변화 등)
    * 데이터의 범위 차이가 유지됨
    * 데이터 범위의 제한이 생김
    * min-max scaling 등
