# 실습 T-test
* 수도권과 비수도권의 강좌정원수 간의 차이가 통계적으로 유의한지 알아보세요

```python
df['센터명'].unique()
```

```python
from scipy import stats
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
def check_loc(x):
    if '서울' in x or '인천' in x:
        return '수도권'
    else:
        return '비수도권'
```

```python
df['센터명'].apply(lambda x: check_loc(x))
```

```python
df['수도권'] = df['센터명'].apply(lambda x: check_loc(x))
```

```python
df[df['수도권'] == '수도권']
```

```python
df[df['수도권'] != '수도권']
```

```python
cap = df[df['수도권'] == '수도권']['강좌정원수']
non_cap = df[df['수도권'] != '수도권']['강좌정원수']
```

```python
cap
```

```python
non_cap
```

```python
stats.ks_2samp(cap, non_cap) # pvalue가 0.05보다 넘어서 결과만 보면 둘은 같은 통계에서 나온 것이 아님.
# 하지만 cap 과 non_cap은 같은 집단에서 나왔으므로 통계가 항상 정확한 것은 아니라는 것을 알 수 있음.
```

```python
stats.wilcoxon(cap, non_cap) # 같은 길이가 아니라서 안됨
```

```python
len(cap)
```

```python
len(non_cap)
```

```python
stats.ttest_ind(cap, non_cap, equal_var = True)
```

```python
stats.ttest_ind(cap, non_cap, equal_var = False)
```

```python
sns.displot(cap)
plt.show()
```

```python
sns.displot(non_cap)
plt.show()
```

```python
sns.displot(df['강좌정원수'])
plt.show()
```

```python
# 정규성, 등분산성

                     O등분산성x
    
O     ttest_ind          |   ttest_ind
      equal_var = True  |  equal_var = False
                         |
정규성 --------------- |---------
                         |
X                        |
                         |
                   wilcoxon test

분산성은 같은 모집단에서 나왔으면 같음.
다른 모집단일 경우 같을 수도 있고 다를 수도 있음.

일단 ks테스트를 해보고 분산성을 판단한 후 시행

wilcoxon test은 두 비교단의 길이가 같아야 하고 ttest_ind은 안같아도 됨
```
