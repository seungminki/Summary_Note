이 테스트의 귀무 가설은 모집단이 정규 분포를 따른다는 것입니다  
대부분의 통계적 유의성 테스트와 마찬가지로 샘플 크기가 충분히 크면 이 테스트는 귀무가설에서 아주 사소한 일탈도 감지할 수 있습니다.(즉, 통계적으로 유의미한 요과가 있을 수 있지만 너무 작아서 실질적인 의미가 없을 수 있음)

정규성(normality) : 어떠한 분포가 정규분포를 따르는가

등분산성(homogeneity of variance) : 분산이 동일한가  
Shapiro-Wilk test(샤피로 윌크 검정)의 귀무가설과 대립가설은 다음과 같습니다.

H0 (귀무가설) : 표본의 모집단이 정규분포를 따른다.

vs.
