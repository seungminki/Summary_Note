# 4. Apply_Map
---

```python
import pandas as pd 
df =pd.read_csv('https://raw.githubusercontent.com/Datamanim/pandas/main/BankChurnersUp.csv',index_col=0)
df.shape
# shape 함수는 형태를 보여줌
```

```python
df['Income_Category'].value_counts()
```

```python
dic = {
    'Unknown'        : 'N',
    'Less than $40K' : 'a',
    '$40K - $60K'    : 'b',
    '$60K - $80K'    : 'c',
    '$80K - $120K'   : 'd',
    '$120K +'        : 'e'   
}
```

```python
df['newIncome'] = df['Income_Category'].map(lambda x: dic[x])
# 조건 대입
```

```python
df['newIncome']  =df['Income_Category'].map(dic)
# 위와 같다
```

```python
df['newIncome']
```

```python
def changeCategory(x):
    if x =='Unknown':
        return 'N'
    elif x =='Less than $40K':
        return 'a'
    elif x =='$40K - $60K':   
        return 'b'
    elif x =='$60K - $80K':    
        return 'c'
    elif x =='$80K - $120K':   
        return 'd'
    elif x =='$120K +' :     
        return 'e' 

df['newIncome'] = df['Income_Category'].apply(changeCategory)

df['newIncome'] 
# 위와 같다
```

```python
df['AgeState'] = df['Customer_Age'].apply(lambda x: x//10 * 10)
df['AgeState'].value_counts().sort_index()
```

```python
df['newEduLevel'] = df['Education_Level'].map(lambda x: 1 if 'Graduate' in x else 0)
Ans = df['newEduLevel'].value_counts()
```

```python
df['Education_Level'].value_counts()
```

```python
df['newEduLevel'].value_counts()
# if 'Graduate' == 1
# NOT 0
```

```python
df['newLimit'] = df['Credit_Limit'].map(lambda x: 1 if x >= 4500 else 0)
df['newLimit'].value_counts()
```

```python
def change_credit(x):
    if x >= 4500:
        return 1
    else:
        return 0
    
df['newLimit'] = df['Credit_Limit'].map(change_credit)
```

```python
df['newLimit'].value_counts()
```

```python
def check(x):
    if x['Marital_Status'] == 'Married' and x['Card_Category'] == 'Platinum' :
        return 1
    else:
        return 0
    
df['newState'] = df.apply(check, axis = 1)
```

```python
df['newState'].value_counts()
```

```python
def ChangeGender(x):
    if x == 'M':
        return 'male'
    else:
        return 'female'

df['Gender'] = df['Gender'].apply(ChangeGender)
```

```python
df['Gender'].value_counts()
```

```python
df['Gender'] = df['Gender'].apply(lambda x: 'male' if x == 'M' else 'female')
df['Gender'].value_counts()
```
